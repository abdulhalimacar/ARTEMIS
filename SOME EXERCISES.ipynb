```python
import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib 
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("tips")
df.sample()
```

```python
sns.scatterplot(x=df["tip"], y=df["total_bill"])

plt.plot([1,10], [3.07,50.81], color="red");
```

```python
df.total_bill.max()
```

```python
sns.lmplot(x="tip", y="total_bill", data=df);
```

```python
#There is a strong relationship between total_bill and tip. More spending means more tips based on the plot below.
sns.scatterplot(x=df["total_bill"], y=df["tip"], hue=df["sex"], alpha=0.9);
```

```python
fig,ax =plt.subplots(figsize=(12,8))
ax=sns.countplot(x=df["smoker"], hue=df["sex"])

for i in ax.patches:
  ax.annotate(i.get_height(), (i.get_x()+0.17, i.get_height()), color="red")

plt.legend(loc=(1,0));
```
